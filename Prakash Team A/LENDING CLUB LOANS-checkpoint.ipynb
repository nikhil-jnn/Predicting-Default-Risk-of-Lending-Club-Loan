```python
rejected_df=pd.read_csv("rejected_2007_to_2018Q4.csv")
rejected_df.head()
```

```python
rejected_df.info()
```

```python
rejected_df.shape
```

```python
rejected_df['Risk_Score'].fillna((rejected_df['Risk_Score'].mean()), inplace=True)
```

```python
rejected_df.isnull().sum(axis=0)
```

```python
rejected_df.info(verbose=True, null_counts=True)
```

```python
import numpy as np
```

```python
cateogry_columns=rejected_df.select_dtypes(include=['object']).columns.tolist()
integer_columns=rejected_df.select_dtypes(include=['int64','float64']).columns.tolist()

for column in rejected_df:
    if rejected_df[column].isnull().any():
        if(column in cateogry_columns):
            rejected_df[column]=rejected_df[column].fillna(rejected_df[column].mode()[0])
        else:
            rejected_df[column]=rejected_df[column].fillna(rejected_df[column].mean)
```

```python
rejected_df.isnull().sum(axis =0)
```

```python
combined_dataset = accept_df.join(rejected_df)
```

```python
combined_dataset.shape
```

```python
loan = combined_dataset['loan_status'].value_counts()
```

# Data Visulaization


```python
sns.distplot(accept_df['loan_amnt'])
plt.gcf().set_size_inches(14,6)
plt.title('Distribution of Loan Amount', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 10)
plt.ylabel('Frequency', fontsize=10)
plt.tight_layout()
```

```python
plt.figure(figsize=(14,8))

sns.barplot(y = loan.index, x = loan,color = 'purple')
plt.title('Loan Staus Values for Lending Club',fontsize=15)
plt.ylabel('Loan Status', fontsize=10)
plt.xlabel('Count',fontsize=10)
```

```python
gr = combined_dataset['grade'].value_counts()
```

```python
plt.figure(figsize=(14,8))

sns.barplot(y=gr.index, x=gr)
plt.title('Loan Grades', fontsize=30)
plt.ylabel('Loan Grade', fontsize=15)
plt.xlabel('Count', fontsize=15)
```

```python
plt.figure(figsize=(30,12))
sns.boxplot(x="installment", y="grade", data=combined_dataset, palette="husl")
plt.title('Box Plot of Installment value vs. Loan Grade', fontsize=25)
plt.xlabel('Installment', fontsize=15)
plt.ylabel('Loan Grade', fontsize=15)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
# Plot the borrower's Loan Amount compared to Loan Purpose
plt.figure(figsize=(30,25))
sns.boxplot(x="loan_amnt", y="purpose", data=combined_dataset, palette="husl")
plt.title('Box Plot of Loan Amount vs. Loan Purpose', fontsize=25)
plt.xlabel('Loan Amount', fontsize=20)
plt.ylabel('Loan Purpose', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
corr = combined_dataset.corr()
# Plot the heatmap
plt.figure(figsize=(14,14))
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,
        cmap='coolwarm')
```
